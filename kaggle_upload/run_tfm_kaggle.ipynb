```python
# === Celda 1: setup del entorno ===
import subprocess, sys, os, shutil, glob

# Nombre (slug) del dataset que subiste. Si Kaggle lo renombró, mira en /kaggle/input.
DATASET_SLUG = "tfm-idrecovery"

# Localiza la carpeta 'tfm' dentro del input (robusto al slug exacto)
candidates = glob.glob(f"/kaggle/input/{DATASET_SLUG}/tfm") + glob.glob("/kaggle/input/*/tfm")
assert candidates, "No encuentro la carpeta 'tfm' en /kaggle/input. ¿Añadiste el dataset?"
src = candidates[0]
print("Código encontrado en:", src)

# Copia a un directorio escribible (los input/ son de solo lectura)
shutil.copytree(src, "/kaggle/working/tfm", dirs_exist_ok=True)
os.chdir("/kaggle/working/tfm")
sys.path.insert(0, "/kaggle/working/tfm")
print("Directorio de trabajo:", os.getcwd())
```

```python
# === Celda 2: dependencias ===
# EDAspy y psutil están en PyPI. pysmile es el wheel de BayesFusion:
# si 'pip install pysmile' falla, sube también el .whl al dataset y
# cambia la línea de pysmile por el path al wheel.
for pkg in ["EDAspy", "psutil", "pysmile"]:
    print(f"--- instalando {pkg} ---")
    subprocess.run([sys.executable, "-m", "pip", "install", pkg, "-q"])

# Comprobación rápida de que pysmile + licencia cargan
import pysmile_license, pysmile
print("pysmile OK")
```

```python
# === Celda 3: lanzar el grid search ===
# Se ejecuta el .py tal cual (sin riesgo de indentación al pegar).
# Cambia --optimizer_type por: umda | egna | emna | keda
subprocess.run([sys.executable, "grid_search.py", "--optimizer_type", "umda"], check=True)
```
